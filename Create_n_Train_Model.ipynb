**getting the popular place info from the google api through code**

```python
import urllib, json
import pprint

URL2 = "https://maps.googleapis.com/maps/api/place/textsearch/json?query=los+angeles+point+of+interest&language=en&key=AIzaSyBIu7cZl56v5JgQw1mMXeMlm7k2DukNaWI"

googleResponse = urllib.urlopen(URL2)
jsonResponse = json.loads(googleResponse.read())
lst_data = jsonResponse['results']
```

**extracting the popular place list from json**

```python
poi_list = []
for data in lst_data:
    poi_list.append(data['name'])
poi_list 
```

** Searching and tagging each tweet with popular_spot name found in text, adding column to the df with that popular_spot name**

```python
def search_str(text):
    text = str(text)
    for poi in poi_list:
        if poi in text.decode('utf-8').strip():
            #print poi
            return poi
        
df['popular_spot'] = df['text'].apply(search_str)
```

```python
df.to_csv("temp.csv")
```

** Extract Sentiment model saved and predict for each tweet**

```python
from sklearn.externals import joblib
filename = 'sentiment_model.sav'
clf = joblib.load(filename)
filename = 'tfidf_vectorizer.sav'
tv = joblib.load(filename)
```

```python
df.dropna(subset=['text'])
```

```python
pd.isnull(df['text']).to_csv("null_text.csv")
df.dtypes
```

```python
df.text.astype(str)
```

```python
tweet_tfidf = tv.transform(df['text'].astype(str))##.values.astype('U'))
```

```python
tweet_pred = clf.predict(tweet_tfidf)
df["sent_pred"] = tweet_pred
df.shape
```

```python
df.to_csv("output_got_LA_Dec2017_sentiment_prediction.csv")
```

**Extracting the total tweets and adding total favorites done per popular spots . We can even total the retweets happening for tweets per popular_spot**

```python

testdf = df.groupby('popular_spot')
#df_extract = testdf.agg({'popular_spot':np.size, 'favorites':np.sum, 'retweets':np.sum, 'sent_pred':np.mean, 'mentions':np.sum, 'hashtags':np.sum })
df_extract = testdf.agg({'popular_spot':np.size, 'favorites':np.sum, 'retweets':np.sum,'mentions':np.count_nonzero,'hashtags':np.count_nonzero, 'sent_pred':np.mean })
```
